```python
d1=pd.read_csv('https://query.data.world/s/wrxyalft7tm2qlg7soxjv75tw5a26o')
d2=pd.read_csv('https://query.data.world/s/kxh2l22bf3id3uj2uutljvwzcbfhv6')
d3=pd.read_csv('https://query.data.world/s/mfg4mszxwdgs4qw7m2kjmqq5wpdw7x')
d4=pd.read_csv('https://query.data.world/s/bbh4zuwcfl6ugi2iahvh5m74sqrbik')
```

```python
d4.columns=['NAME','JOBTITLE','DEPTID','DESCR','HIRE_DT','GROSS','ANNUAL_RT']
```

```python
data_frames = [d1, d2, d3, d4]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['NAME','JOBTITLE','DEPTID','DESCR','HIRE_DT'],
                                            how='outer'), data_frames).fillna(0)
```

```python
df_merged.columns=['NAME','JOBTITLE','DEPTID','DESCR','HIRE_DT','GROSS-2019','ANNUAL_RT-2019','GROSS-2018','ANNUAL_RT-2018','GROSS-2017','ANNUAL_RT-2017','GROSS-2016','ANNUAL_RT-2016']
df_merged
```

df_merged.to_excel("op.xlsx")

```python
final_merge = df_merged.groupby(['NAME','JOBTITLE','DEPTID','DESCR','HIRE_DT','GROSS-2019','ANNUAL_RT-2019','GROSS-2018','ANNUAL_RT-2018','GROSS-2017','ANNUAL_RT-2017','GROSS-2016','ANNUAL_RT-2016'],as_index=False).sum()
```
