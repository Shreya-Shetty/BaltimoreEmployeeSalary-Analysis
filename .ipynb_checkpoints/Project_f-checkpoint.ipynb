Reading the files and converting into individual dataframes using pandas.

```python
d1=pd.read_csv('https://query.data.world/s/wrxyalft7tm2qlg7soxjv75tw5a26o')
d2=pd.read_csv('https://query.data.world/s/kxh2l22bf3id3uj2uutljvwzcbfhv6')
d3=pd.read_csv('https://query.data.world/s/mfg4mszxwdgs4qw7m2kjmqq5wpdw7x')
d4=pd.read_csv('https://query.data.world/s/bbh4zuwcfl6ugi2iahvh5m74sqrbik')
```

```python
d4.columns=['NAME','JOBTITLE','DEPTID','DESCR','HIRE_DT','GROSS','ANNUAL_RT']
```

```python
data_frames = [d1, d2, d3, d4]
```

Merging the dataframes into one for furthur analysis.

```python
df = reduce(lambda  left,right: pd.merge(left,right,on=['NAME'],
                                            how='outer'), data_frames).fillna(0)
```

```python
df_merged=df.drop_duplicates(subset='NAME')
```

# Data Cleaning

We can see that the columns 'DESCR'(description) is just additional information for the 'JOBTITLE'. So, we can drop these columns, to save space and optimize.

```python
df_merged=df_merged.drop(['DESCR_x','DESCR_y'],axis=1)
```

Renaming the column names to avoid confusion between the yearly data.

```python
df_merged.columns=['NAME','JOBTITLE_2019','DEPTID_2019','HIRE_DT_2019','ANNUAL_RT_2019','GROSS_2019','JOBTITLE_2018','DEPTID_2018','HIRE_DT_2018','ANNUAL_RT_2018','GROSS_2018','JOBTITLE_2017','DEPTID_2017','HIRE_DT_2017','ANNUAL_RT_2017','GROSS_2017','JOBTITLE_2016','DEPTID_2016','HIRE_DT_2016','ANNUAL_RT_2016','GROSS_2016']
```

```python
df_merged = df_merged.astype({"NAME":str,"JOBTITLE_2019":str,'DEPTID_2019':str,"HIRE_DT_2019":str,"ANNUAL_RT_2019":str,"GROSS_2019":str,"JOBTITLE_2018":str,'DEPTID_2018':str,"HIRE_DT_2018":str,"ANNUAL_RT_2018":str,"GROSS_2018":str,"JOBTITLE_2017":str,'DEPTID_2017':str,"HIRE_DT_2017":str,"ANNUAL_RT_2017":str,"GROSS_2017":str,"JOBTITLE_2016":str,'DEPTID_2016':str,"HIRE_DT_2016":str,"ANNUAL_RT_2016":str,"GROSS_2016":str})
```

Editing the salaries by removing the '$' symbol to access it as an integer.

```python
df_merged.ANNUAL_RT_2018=df_merged.ANNUAL_RT_2018.apply(lambda x: x.replace('$',''))
df_merged.ANNUAL_RT_2017=df_merged.ANNUAL_RT_2017.apply(lambda x: x.replace('$',''))
df_merged.ANNUAL_RT_2016=df_merged.ANNUAL_RT_2016.apply(lambda x: x.replace('$',''))
df_merged.GROSS_2018=df_merged.GROSS_2018.apply(lambda x: x.replace('$',''))
df_merged.GROSS_2017=df_merged.GROSS_2017.apply(lambda x: x.replace('$',''))
df_merged.GROSS_2016=df_merged.GROSS_2016.apply(lambda x: x.replace('$',''))
```

```python
df_merged = df_merged.astype({"ANNUAL_RT_2019":'float64',"GROSS_2019":'float64',"ANNUAL_RT_2018":'float64',"GROSS_2018":'float64',"ANNUAL_RT_2017":'float64',"GROSS_2017":'float64',"ANNUAL_RT_2016":'float64',"GROSS_2016":'float64'})
```

Extracting the year from the hiring date which can be used for analysis. 

```python
df_merged['converted_date']=pd.to_datetime(df_merged['HIRE_DT_2019'],format='%m/%d/%Y %H:%M:%S %p',errors="coerce")
```

```python
df_merged['year'] = df_merged['converted_date'].dt.year
```

# Data analyzing

Hypothesis 1: The number of salary hikes will be more than deductions, when 2016 to 2019 datasets are compared.

```python
df_merged['sal_diff'] = df_merged['ANNUAL_RT_2019']-df_merged['ANNUAL_RT_2016'] #Calculating the difference
```

```python
#Displaying the data for verification purpose
df_merged['sal_diff']
```

I want to find the difference in salary only of the person is employeed in both the years 2016 and 2019. With the help of the 'if' conditional looping, I am iterating through each row using 'iterrows' method.

```python
inc=0
dec=0
for index, row in df_merged.iterrows():
    if(row[4]!=0 and row[19]!=0 and row[23]>0):
        inc=inc+1
    elif(row[4]!=0 and row[19]!=0 and row[23]<0):
        dec=dec+1   

print("Number of hikes: "+str(inc))
print("Number of reduced salaries: "+str(dec))
```

Hypothesis 1 proved correct. The number of increase in salaries is more than number of decreased salaries.

Hypothesis 2: Employees in Police department might have the highest gross income.

```python
col_list=['GROSS_2019','GROSS_2018','GROSS_2017','GROSS_2016'] #Selecting only the columns having gross income
max_val=(df_merged[col_list].max()).max()
# 'df_merged[col_list].max()' gives the max income in each column
for index,row in df_merged[col_list].iterrows():
    if(row[0]== max_val or row[1]== max_val or row[2]== max_val or row[3] == max_val):
        i=index #saving the index of the row having max value
for index,row in df_merged.iterrows():
    if(index==i): #comparing the required index with all the indexes
        print(row)
```

Hypothesis 2 proved right, as the max gross income,$260775, is for the police department, job titled 'Police Sergeant EID'.

Question 1: Which is the highest paid profession?
<br>
This question is similar to Hypothesis 2, but here I am including the annual incomes too, to check if there's any employee earning annual income more than the police department.

```python
col_list=['GROSS_2019','GROSS_2018','GROSS_2017','GROSS_2016','ANNUAL_RT_2016','ANNUAL_RT_2017','ANNUAL_RT_2018','ANNUAL_RT_2019']
index=(df_merged[col_list].idxmax()).max()
for i ,row in df_merged.iterrows():
    if(index==i):
        print(row[1])
```

It can be concluded that, POLICE department employees earn the most(in the taken dataset).

Question 2: Is the same profession paid highest in all the four years?

Here, I am defining a function which retrieves the profession of the maximum salary paid.

```python
def find_max(l1):
    for x in l1:
        index=(df_merged[x].idxmax()).max() #saving the index value of the max salary in the given column
        for i ,row in df_merged.iterrows():
            if(index==i):
                print(x+" : "+row[1])
```

```python
col_list=['ANNUAL_RT_2016','ANNUAL_RT_2017','ANNUAL_RT_2018','ANNUAL_RT_2019']
find_max(col_list)
```

From the above output, it is clear that, same profession is not the highest paid always!

I can see that 'Executive Director III' is paid highest twice, plotting a graph with all the salaries of this profession for the years 2016-2017, to find the range of their salary.

```python
plt.plot(df_merged[df_merged['JOBTITLE_2017']=="Executive Director III"]['ANNUAL_RT_2017'],marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=3)
plt.plot(df_merged[df_merged['JOBTITLE_2016']=="Executive Director III"]['ANNUAL_RT_2016'],marker='o', markerfacecolor='green', markersize=12, color='skyblue', linewidth=4)
plt.ylabel('Employee salary',fontsize=12)
plt.xlabel('Serial number(in the dataframe)',fontsize=12)
plt.title('"Executive Director III" salaries in 2016 and 2017')
```

From the graph: Not every Executive Director is highly paid. There's a difference of $70,000 seen. 

Question 3: Which profession has a drastic reduce in remuneration?

To find change in the remunerations, I have created a dataframe which has the entries of employees who have worked in the same profession all the years. The new dataframe shall have the values of the employee name, salary in the year 2019, salary in the year 2016, using which difference can be found.

```python
s=df_merged.groupby('JOBTITLE_2019')
```

```python
job19=pd.DataFrame({'sal_2019' : s['ANNUAL_RT_2019'].mean()}).reset_index()
job19.columns=['job_desc','sal_2019']
```

```python
s1=df_merged.groupby('JOBTITLE_2016')
```

```python
job16=pd.DataFrame({'sal_2016' : s1['ANNUAL_RT_2016'].mean()}).reset_index()
job16.columns=['job_desc','sal_2016']
```

```python
res=pd.merge(job19,job16,how='outer')
res=res.fillna(0)
l=[]
for i in range(1,len(res)):
    if not (res['sal_2019'][i]==0.000000 or res['sal_2016'][i]==0.000000):
        l.append([res['job_desc'][i],res['sal_2019'][i],res['sal_2016'][i],res['sal_2016'][i]-res['sal_2019'][i]])
l
```

```python
find_max=[]
sample=0
for i in l:
    if(i[3]>0): #to store all the rows with reduced remuneration
        find_max.append(i)
sample=find_max[0]
for i in find_max: #finding the maximum reduced remuneration
    if(i[3]>=sample[3]):
        for j in range(0,4):
            sample[j]=i[j]
        
print(sample)
```

An employee working as a 'Chief of Engineering' had a drastic reduction in his remuneration, with an amount of $17,700.

Checking the years, where number of recruitments was more than 500.

```python
year=df_merged.groupby('year')['NAME'].count() #grouping the data w.r.t year of hiring
year_df=pd.DataFrame(year).reset_index()
year_df.columns=['year','count']
year_df = year_df.astype({"year":int,"count":int})
for index,row in year_df.iterrows(): #iterating through rows to find the 
    if (row['count']<500):
        year_df.drop(index, inplace=True) #dropping the rows with less that 500
#Using plotly for better visualization        
fig = px.bar(year_df, x='year', y='count')

fig.update_layout(
    autosize=False,
    width=700,
    height=345,
    title_text="More than 500 recruitments"
)

fig.show()
```
